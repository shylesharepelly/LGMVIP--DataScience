```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from skimage import io
```

dataset link: https://i.pinimg.com/originals/c6/e7/df/c6e7dfb7c951eed5cb516b56a4751632.jpg

```python
image = io.imread("https://i.pinimg.com/originals/c6/e7/df/c6e7dfb7c951eed5cb516b56a4751632.jpg")
```

```python
cv2_imshow(image)
```

```python
gray_filter = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_filter)
```

```python
inverted_filter = cv2.bitwise_not(gray_filter)
cv2_imshow(inverted_filter)
```

```python
image_smoothing = cv2.GaussianBlur(inverted_filter, (21, 21), sigmaX=0, sigmaY=0)
cv2_imshow(image_smoothing)
```

```python
def pencil_sketch(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch = pencil_sketch(gray_filter, image_smoothing)
cv2_imshow(pencil_sketch)
```

CONCLUSION:

Image is converted into a sketch using python.
